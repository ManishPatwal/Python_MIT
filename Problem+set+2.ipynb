# Problem 1

G&A produces and distributes personal care products. The goods are manufactured at 6 different plants and then sent to different distribution centers throughout the United States. 
Some distribution centersâ€™ managers have been reporting the occurrence of stock-outs in the last few months. 

You were tasked to analyze inventory on-hand data for each SKU at each distribution center to quantify how big the problem is. To perform this task, you were given two Excel files that contain four columns described below:

- date: date
- dc: distribution center id
- sku: SKU id
- inventory: inventory on-hand (in cases)

In order to complete your analysis, perform the following tasks:

- Read the 2 Excel files (dc_inventory(1).xlsx and dc_inventory(2).xlsx)
- Concatenate the 2 inventory datasets to create a single dataset. The resulting DataFrame should contain 118,650 rows
- Sort the values by DC, SKU and Date in ascending order
- Identify how many unique distribution centers are included in the dataset
- Identify how many unique SKUs are included in the dataset
- Add a new column named 'oos' to identify out-of-stocks. If inventory on-hand > 0, then the value of this column should be 0, otherwise 1. Hint: use the lambda function
- Create a new DataFrame computing the number of days with out-of-stocks by DC and SKU. The columns of this new DataFrame should be 'dc', 'sku', and 'oos_days'. The number of days with out-of-stocks (oos_days) should be the sum of 'oos' for each SKU at ecah DC. Hint: use groupby
- In the new DataFrame, remove those rows with 350 days of stockouts
- Identify the DC and SKU with the highest number of days with stockouts
- Save the resulting DataFrame as an Excel file

```python
# Write your code here
import pandas as pd
import numpy as np
Inventory_1=pd.read_excel("C:/Users/manis/Desktop/MIT_Python/dc_inventory(1).xlsx")
Inventory_2=pd.read_excel("C:/Users/manis/Desktop/MIT_Python/dc_inventory(2).xlsx")

#Concatenate the 2 inventory datasets to create a single dataset. The resulting DataFrame should contain 118,650 rows
Inventory=pd.concat([Inventory_1,Inventory_2])
Inventory.shape #checks the shape- Data has 118560 rows and 4 columns
Inventory.head()
#Sort the values by DC, SKU and Date in ascending order
Inventory_data=Inventory.sort_values(by=['dc', 'sku','date'])
Inventory_data.head()

#Identify how many unique distribution centers are included in the dataset
unique_dc=Inventory_data['dc'].unique()
print('Unique Dc are',len(unique_dc)) 
# 45 Unique DC

#Identify how many unique SKUs are included in the dataset

unique_sku=Inventory_data['sku'].unique()
print('Unique sku are',len(unique_sku)) 
#  Unique skus are 121

#Add a new column named 'oos' to identify out-of-stocks. 
#If inventory on-hand > 0, then the value of this column should be 0, otherwise 1. Hint: use the lambda function


Inventory_data['oos'] = list(map(lambda x:1 if x>0 else 0 , Inventory_data['inventory'])) 
#lets Validate our calculation
Inventory_data.tail()
#the above tail only has zero inventory values .
#lets check and validdate through somewhere inventory is>0
Inventory_data[Inventory_data.inventory>0].tail()
#validated the lambda function


# Create a new DataFrame computing the number of days with out-of-stocks by DC and SKU. 
#The columns of this new DataFrame should be 'dc', 'sku', and 'oos_days'. 
#The number of days with out-of-stocks (oos_days) should be the sum of 'oos' for each SKU at ecah DC. Hint: use groupby
groupd_data=Inventory_data.groupby(['dc','sku'])['oos'].sum()
groupd_data = groupd_data.to_frame().reset_index()
type(groupd_data)
groupd_data.rename(columns={'oos':'oos_days'}, inplace=True) #change name of oos to oos_days

groupd_data.head(50)
#groupd_data.columns
groupd_data.shape #current shape is (339,3)


#In the new DataFrame, remove those rows with 350 days of stockouts
new_data=groupd_data[groupd_data['oos_days']!=350]
new_data.head()
new_data.shape #new shape is (333,3) i.e 6 values of oos_days were equal to 350 and ar dropped.



#Identify the DC and SKU with the highest number of days with stockouts

## lets check for DC first
dc_oos=  new_data.groupby(['dc'])['oos_days'].sum()
dc_oos = dc_oos.to_frame().reset_index()
dc_oos.head()
dc_oos.loc[dc_oos['oos_days'].idxmax()] #This gives dc 520 as having maximum Out of Stock days of 2284 days

## lets check for DC first
dc_oos=  new_data.groupby(['dc'])['oos_days'].sum()
dc_oos = dc_oos.to_frame().reset_index()
dc_oos.head()
dc_oos.loc[dc_oos['oos_days'].idxmax()] #This gives dc 520 as having maximum Out of Stock days of 2284 days
print("DC with highest out of stock days is",dc_oos.loc[dc_oos['oos_days'].idxmax()])
##SKU with the highest number of days with stockouts
sku_oos=  new_data.groupby(['sku'])['oos_days'].sum()
sku_oos = sku_oos.to_frame().reset_index()
sku_oos.head()
sku_oos.loc[sku_oos['oos_days'].idxmax()] #This gives dc 520 as having maximum Out of Stock days of 2284 days
print("SKU with highest out of stock days is",sku_oos.loc[sku_oos['oos_days'].idxmax()])
```

# Problem 2

G&A suspects that stock-outs are caused by missed deliveries. To prove this hypothesis, you were tasked to analyze shipment data over the period of 1 year. You were handed an Excel file containing the information detailed below:

- date: date
- plant: plant's name (origin)
- dc: distribution center id (destination)
- sku: SKU id
- ordered: number of cases ordered 
- delivered: number of cases delivered 

In order to complete your analysis, perform the following tasks:

- Read the Excel file that contains shipment data from plants to DCs (dc_shipments.xlsx)
- Sort the values by DC, SKU, Plant, and Date in ascending order
- Create a new column (missed) to compute the number of cases missed (ordered - delivered)
- Create a new DataFrame computing the number of cases missed by Plant. Hint: use groupby
- Identify the top 3 Plants with the largest number of cases missed
- Create a new DataFrame computing the number of cases missed by DC. Hint: use groupby
- Identify the top 5 DCs with the largest number of cases missed
- Create a new DataFrame computing the number of cases missed by SKU. Hint: use groupby
- Identify the top 5 SKUs with the largest number of cases missed

```python
# Write your code here
import pandas as pd
import numpy as np

#Read the Excel file that contains shipment data from plants to DCs (dc_shipments.xlsx)
dc_shipments=pd.read_excel("C:/Users/manis/Desktop/MIT_Python/dc_shipments.xlsx")
print(dc_shipments.head())

#Sort the values by DC, SKU, Plant, and Date in ascending order
dc_shipments=dc_shipments.sort_values(by=['dc', 'sku','plant','date'])
print(dc_shipments.head())

#Create a new column (missed) to compute the number of cases missed (ordered - delivered)
dc_shipments['missed']=dc_shipments['ordered']-dc_shipments['delivered']
print(dc_shipments.head(2))
print((dc_shipments[dc_shipments['missed']>0]).head(2)) # Verifies for missed >0

#Create a new DataFrame computing the number of cases missed by Plant. Hint: use groupby
Plant_Case_missed=dc_shipments.groupby(['plant'])['missed'].sum()
Plant_Case_missed = Plant_Case_missed.to_frame().reset_index()

#Identify the top 3 Plants with the largest number of cases missed
print('The top 3 plants for cases missed are:\n',Plant_Case_missed.sort_values(by=['missed'],ascending=False).head(3))


#Create a new DataFrame computing the number of cases missed by DC. Hint: use groupby
dc_Case_missed=dc_shipments.groupby(['dc'])['missed'].sum()
dc_Case_missed = dc_Case_missed.to_frame().reset_index()

#Identify the top 5 DCs with the largest number of cases missed

print('The top 5 dc for cases missed are:\n',dc_Case_missed.sort_values(by=['missed'],ascending=False).head(5))


#Create a new DataFrame computing the number of cases missed by SKU. Hint: use groupby
sku_Case_missed=dc_shipments.groupby(['sku'])['missed'].sum()
sku_Case_missed=sku_Case_missed.to_frame().reset_index()

#Identify the top 5 SKUs with the largest number of cases missed
print('The top 5 SKUs for cases missed are:\n',sku_Case_missed.sort_values(by=['missed'],ascending=False).head(5))
```
