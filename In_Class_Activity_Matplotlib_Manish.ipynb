```python
#Activity 1: Simple plots

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

month = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
demand = [72, 110, 117, 172, 76, 112, 130, 194, 78, 119, 130, 201] 
deseasonalized_demand = [117, 122, 117, 117, 123, 124, 130, 131, 127, 132, 128, 137]
plt.plot(month,demand,'b-o',month,deseasonalized_demand,'r--^')
plt.title("Demand over time")
plt.xlabel("Month")
plt.ylabel("Units")
plt.show()
#plt.plot(n,25*n,'r--',n,n**2,'bs',n,n**3,'g^')
# Write your code here
```

```python
#Activity 2: Line plots
 # Write your code here
import pandas as pd
data=pd.read_excel("C:\\Users\manis\\Desktop\\MIT_Python\\CO2_2017_daily.xlsx")

#Based on the CO2 emissions file (CO2_2017_daily), 
#create a line plot to show the average CO2 emissions per day (Ave CO2 emissions (KgCO2/Ton-Km))
#plt.plot(month,demand,'b-o',month,deseasonalized_demand,'r--^')
data[["Day","Ave CO2 emissions (KgCO2/Ton-Km)"]].plot(style="g-",title="Average of CO2 emissions per day",xlim=[0,90],ylim=[0.055,0.075])
plt.xlabel('Day')
plt.ylabel("Daily CO2 emissions (KgCO2/Ton-Km)")
   
```

```python
#Activity 3: Bar plots

data = pd.read_excel("C:\\Users\manis\\Desktop\\MIT_Python\\CO2_2017_daily.xlsx")
data_month = data['Ave CO2 emissions (KgCO2/Ton-Km)'].groupby(data['Month']).mean().reset_index(name='Ave CO2 emissions per month')
#data_month['Month']=data_month['Month'].astype('categorical')
# Write your code here
print(data_month)
data_month["Ave CO2 emissions per month"].plot(style="g",kind="bar",title="Average of CO2 emissions per day")
plt.xlabel("Month")
plt.ylabel('CO2 emissions (KgCO2/Ton-Km)')
title="Average of CO2 emissions per month"
```
