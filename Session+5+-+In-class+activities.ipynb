# Activity 1: Simple plots

Given two time series (demand and deseasonalized demand), create a single chart with the following characteristics:

- Demand: Blue solid line with a circle marker
- Deseasonalized demand: Red dashed line with a triangle-up marker
- Title: “Demand over time” 
- x and y labels: “Month” and “Units”, respectively

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

month = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
demand = [72, 110, 117, 172, 76, 112, 130, 194, 78, 119, 130, 201] 
deseasonalized_demand = [117, 122, 117, 117, 123, 124, 130, 131, 127, 132, 128, 137]
plt.plot(month,demand,'b-o',month,deseasonalized_demand,'r--^')
plt.title("Demand over time")
plt.xlabel("Month")
plt.ylabel("Units")
plt.show()
#plt.plot(n,25*n,'r--',n,n**2,'bs',n,n**3,'g^')
# Write your code here
```
